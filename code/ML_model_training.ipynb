# model training  
### SVM, NN, Naive Bayes, Random forest, KNN, Logistic Regression

```python
for t in range(1):
    df = pd.read_excel(f'D:/m/new_img/feature.xlsx')
    #df = pd.read_excel(f'D:/m/train{t+1}_std.xlsx')
    train_x = np.array(m_train.iloc[:, 1:-1])
    train_y = np.array(m_train.iloc[:, -1])
    #test_x1 = pd.read_excel(f'D:/m/train{t+1}_std.xlsx')
    test_x = np.array(m_test.iloc[:,1:-1])
    test_y = np.array(m_test.iloc[:, -1])

    TP=[]
    FN=[]
    FP=[]
    TN=[]
    ACC=[]
    TIME=[]
    
    # SVM
    start = time.time()
    C = 1
    opts = dict(C = C, tol = 1e-8, max_iter = int(1e8))
    clf_svm = SVC(kernel='rbf', gamma='auto', **opts)
    clf_svm.fit(train_x, train_y)
    predictions = clf_svm.predict(test_x)
    acc = clf_svm.score(test_x,test_y)
    cm = confusion_matrix(test_y, predictions)
    cm_display = ConfusionMatrixDisplay(cm).plot()
    end = time.time()
    tn, fp, fn, tp = confusion_matrix(test_y, predictions).ravel()
    TP.append(tp)
    FN.append(fn)
    FP.append(fp)
    TN.append(tn)
    ACC.append(acc)
    TIME.append(end-start)

    #NN(logistic)
    start = time.time()
    hidden_layers = (512,)
    activation = 'relu'
    opts = dict(hidden_layer_sizes = hidden_layers, verbose = False, \
    activation = activation, tol = 1e-8, max_iter = int(1e8))
    solver = 'adam' 
    clf_MLP = MLPClassifier(solver = solver, **opts)
    clf_MLP.fit(train_x, train_y)
    predictions = clf_MLP.predict(test_x)
    acc = clf_MLP.score(test_x,test_y)
    end = time.time()
    cm = confusion_matrix(test_y, predictions)
    cm_display = ConfusionMatrixDisplay(cm).plot()
    tn, fp, fn, tp = confusion_matrix(test_y, predictions).ravel()
    TP.append(tp)
    FN.append(fn)
    FP.append(fp)
    TN.append(tn)
    ACC.append(acc)
    TIME.append(end-start)

    # Naive bayes
    start = time.time()
    modelg=GaussianNB()
    modelg.fit(train_x,train_y)
    pred = modelg.predict(test_x)
    acc = modelg.score(test_x,test_y)
    end = time.time()
    cm = confusion_matrix(test_y, pred)
    cm_display = ConfusionMatrixDisplay(cm).plot()
    tn, fp, fn, tp = confusion_matrix(test_y, pred).ravel()
    TP.append(tp)
    FN.append(fn)
    FP.append(fp)
    TN.append(tn)
    ACC.append(acc)
    TIME.append(end-start)

    # Random forest
    start = time.time()
    forest = ensemble.RandomForestClassifier(n_estimators = 150,max_depth=20)
    forest_fit = forest.fit(train_x, train_y)
    test_y_predicted = forest.predict(test_x)
    accuracy = metrics.accuracy_score(test_y, test_y_predicted)
    end = time.time()
    cm = confusion_matrix(test_y, test_y_predicted)
    cm_display = ConfusionMatrixDisplay(cm).plot()
    tn, fp, fn, tp = confusion_matrix(test_y, test_y_predicted).ravel()
    TP.append(tp)
    FN.append(fn)
    FP.append(fp)
    TN.append(tn)
    ACC.append(accuracy)
    TIME.append(end-start)

    # KNN
    start = time.time()
    knn = KNeighborsClassifier(n_neighbors=5,weights = 'distance',metric = 'manhattan')
    knn.fit(train_x, train_y)
    y_pred = knn.predict(test_x)
    acc = metrics.accuracy_score(test_y, y_pred)*100
    end = time.time()
    cm = confusion_matrix(test_y, y_pred)
    cm_display = ConfusionMatrixDisplay(cm).plot()
    tn, fp, fn, tp = confusion_matrix(test_y, y_pred).ravel()
    TP.append(tp)
    FN.append(fn)
    FP.append(fp)
    TN.append(tn)
    ACC.append(acc)
    TIME.append(end-start)
    
    #logistic
    start = time.time()
    x_train_const = tools.add_constant(train_x)
    log_reg_3 = sm.Logit(train_y, x_train_const).fit()
    x_test_const = tools.add_constant(test_x)
    pred_3 = log_reg_3.predict(x_test_const)
    prediction_3 = list(map(round, pred_3))
    end = time.time()
    accuracy = metrics.accuracy_score(test_y, prediction_3)
    tn, fp, fn, tp = confusion_matrix(test_y, prediction_3).ravel()
    TP.append(tp)
    FN.append(fn)
    FP.append(fp)
    TN.append(tn)
    ACC.append(accuracy)
    TIME.append(end-start)

    path = r'D:\m\new_img_result.txt'
    f = open(path, 'a')
    for i in range(len(TP)):
        lines = [str(TP[i]), ' ', str(FN[i]), ' ', str(FP[i]), ' ', str(TN[i]), ' ', str(ACC[i]),' ', str(TIME[i]), ' ']
        f.writelines(lines)

    f.writelines('\n')
    f.close()
    
path = r'D:\m\new_img_result.txt'
f = open(path, 'a')
f.writelines('\n')
f.close()
```
