```python
import pandas as pd
from sklearn import preprocessing
import numpy as np
from sklearn.preprocessing import StandardScaler
data = pd.read_excel("D:/m/merge.xlsx")
print(data)
```

# Normalization

```python
x=data.values
y=x[:,32]
#print(y)
x2=np.delete(x,32,axis=1)
#print(x2[:,3])
mu = x2.mean(axis=0)
std = x2.std(axis=0)
N=data.shape[0]
x_st = (x2-np.tile(mu,(N,1)))/std
print(x_st[:,3])
```

```python
mdata=np.c_[x_st,y]
print(mdata)
```

# Write to Excel

```python
import os
os.chdir('D:/m')
import openpyxl
wb = openpyxl.Workbook()
list_charac = mdata.tolist()
s1 = wb.create_sheet('data')
column_name = ['ae_diff', 'ar_diff','Proportion', 'h_mean' , 'h_std', 'h_skw', 's_mean', 's_std', 's_skw', 'v_mean', 'v_std', 
              'v_skw', 'contrast0', 'contrast45', 'contrast90', 'contrast135', 'dissimilarity', 'dissimilarity45', 'dissimilarity90', 
              'dissimilarity135', 'homogeneity0', 'homogeneity45', 'homogeneity90','homogeneity135', 'energy0', 'energy45',
              'energy90', 'energy135', 'correlation0', 'correlation45', 'correlation90', 'correlation135','y']
s1.append(column_name)

for i in list_charac:
    s1.append(i)      


wb.save('mdata_st.xlsx')
```
